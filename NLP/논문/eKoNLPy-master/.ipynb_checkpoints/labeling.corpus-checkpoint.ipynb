```python
def compare_append(txt):
    txt['Polarity'] = ''
    Hawkish = []
    Dovish = []
    Grey = []
    for i in range(len(txt)):
        temp_date = txt['date'][i]
        if Callrate_labeling.loc[str(temp_date)][0] == 'up':
            Hawkish.append(txt['article'][i])#content로 바꿔주기
            txt['Polarity'][i] = 'H'
        elif Callrate_labeling.loc[str(temp_date)][0] == 'down':
            Dovish.append(txt['article'][i])#content로 바꿔주기
            txt['Polarity'][i]='D'
        else:
            Grey.append(txt['article'][i])#content로 바꿔주기
            txt['Polarity'][i]='N'
    Labeled = txt
    return Labeled, 'Hakish:', Hawkish, 'Dovish: ', Dovish, 'Grey: ',Grey#Labled를 따로 받을 수 있게 해주자
```
