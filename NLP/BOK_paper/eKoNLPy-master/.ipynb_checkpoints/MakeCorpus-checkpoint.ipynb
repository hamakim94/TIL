# Corpus Dictionary csv 데이터 읽기

```python
import pandas as pd
```

```python
CORNEWS = pd.read_csv('news_corpus.csv')
CORBONDS = pd.read_csv('Bond_corpus.csv')
CORBOK = pd.read_csv('BoK_minutes_corpus.csv')
```

```python
COR1 = {}
for i in range(len(CORNEWS)):
    row = CORNEWS.iloc[i]
    COR1[row[0]] = [row[1],row[2],row[3]]
COR1
```

```python
COR2 = {}
for i in range(len(CORBONDS)):
    row = CORBONDS.iloc[i]
    COR2[row[0]] = [row[1],row[2],row[3]]
COR2
```

```python
COR3 = {}
for i in range(len(CORBOK)):
    row = CORBOK.iloc[i]
    COR3[row[0]] = [row[1],row[2],row[3]]
```

```python
COR3
```

```python
from tqdm.notebook import tqdm
for key, value in tqdm(COR2.items()):
    if key in COR1.keys():
        for i in range(3):
            COR1[key][i] += value[i]
    else:
        COR1[key] = value
```

```python
for key, value in tqdm(COR3.items()):
    if key in COR1.keys():
        for i in range(3):
            COR1[key][i] += value[i]
    else:
        COR1[key] = value
```

```python
len(COR1)
```

```python
hawkish = []
dovish = []
for key, value in tqdm(COR1.items()):
    if (value[0]+value[1]+value[2])>14:
        if value[2] !=0:
            if (value[0]/value[2]) >= 1.3:
                hawkish.append(key)
            elif (value[0]/value[2]) < (1/1.3):
                dovish.append(key)
        else:
            hawkish.append(key)
```

```python
dov  =pd.DataFrame(dovish, columns = ['dovish'])
dov.to_csv('dovish.csv')
```

```python
hawk = pd.DataFrame(hawkish, columns = ['hawkish'])
hawk.to_csv('hawkish.csv')
```
