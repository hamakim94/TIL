```python
import pandas as pd
BoK_minutes = pd.read_csv("C:/Users/User/Desktop/BoK_minutes.csv")
BoK_minutes['tokens'] = ''
```

```python
BoK_minutes['contents'].apply(lambda x : x.replace('[','').replace(']','').replace('\'','').split(','))
```

```python
BoK_minutes['contents']=BoK_minutes['contents'].apply(lambda x: x.split('.'))
```

```python
from ekonlpy.sentiment import MPCK
pck= MPCK()
```

```python
BoK_minutes['t']
```

```python
repeat_per=[]
for idx, ff in enumerate(BoK_minutes['contents']):
    token_only=[]
    for jdx, fff in enumerate(ff):
        tokk=[]
        tokens=pck.tokenize(fff)
        if tokens !=[]:
            token_only.append(tokens)
            repeat_per.append([len(list(set(tokens)))/len(tokens),tokens])
    BoK_minutes['tokens'][idx]=token_only
```

```python
for k in BoK_minutes['tokens']:
    for i in k:
        for j in i:
            if j[-3:] == 'XSV':
                i.remove(j)
            else:
                pass
```

```python
BoK_minutes['ngram(token+ngram)'] = ''
for i in range(len(BoK_minutes)):
    for j in BoK_minutes['tokens'][i]:
        BoK_minutes['ngram(token+ngram)'][i] = pck.ngramize(j)
# for b in BoK_minutes['tokens'][3]:
#   a.append(pck.ngramize(b))
```
