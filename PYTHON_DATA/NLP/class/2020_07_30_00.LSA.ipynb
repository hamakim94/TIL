### 토픽 모델링  
- 문서 집합의 추상적인 '주제'를 발견하기 위한 통계적 모델 중 하나
- 텍스트 본뮨의 숨겨진 의미 구조를 발견하기 위해 사용  

만약 개에 대한 문서 : 개, 뼈다귀 많아  
고양이 : 고양이, 야옹  
- 각 토픽을 추출된 키워드들의 분포로 나타냄으로서 텍스트 내의 구조 파악
- 빈도수 기반(BOW)
- 뉴스기사, 주요 이슈 추출, 트렌드 분석, 사회와 시대를 이해하기 위해 의사결정

### 잠재 의미 분석
(LSA, Latent Semantic Analysis)  
의미 : 거리, ngram, 얼마나 가까이 있는지?
- 토픽 모델링 방법중에 하나다
- BoW에 기반한 TF-IDF 빈도로, 의미를 고려하지 못한다는 단점이 있다
- 그래서 co0occurrence 한 벡터를 가정(한 문장 내에서 같이 등장하면 의미가 있다) 매우 중요한 개념!
- TDM내 잠재된 의미를 이끌어 내는 방법


### co-occurence : 단어의 형태가 아닌 의미를 사용한다는 뜻
배라는 단어가 같은 문장에 동사가 타다인지 먹다인지 따라 의미가 달라지게 되는데 -> 이게 매우 중요한 개념

### 1) 직접 구현

```python
import numpy as np
from collections import defaultdict
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.decomposition import randomized_svd

class LSA:
    def __init__(self, doc_ls, topic_num):
        self.doc_ls = doc_ls
        self.topic_num = topic_num
        
        self.term2idx, self.idx2term = self.toIdxDict( ' '.join(doc_ls).split())
        self.doc2idx, self.idx2doc = self.toIdxDict(doc_ls)
        
        self.tdm = self.TDM(doc_ls)
        
        self.U, self.s , self.VT = self.SVD(self.tdm)
        
        self.term_mat = self.TermVectorMatrix(self.U, self.s, topic_num)
        self.doc_mat = self.DocVectorMatrix(self.s, self.VT, topic_num)
        
        self.term_doc_mat = self.TermDocVectorMatrix(self.U, self.s, self.VT, topic_num) 
        
    def toIdxDict(self, ls):
        any2idx = defaultdict(lambda : len(any2idx))
        idx2any = defaultdict()
        
        for item in ls:
            any2idx[item]
            idx2any[any2idx[item]] = item
            
        return any2idx, idx2any   
    
    
    # 리스트내 값을 index로 변환하는 dict과 
    # index를 리스트내 값으로 변환하는 dict
    
    def TDM(self, doc_ls):
        tdm = np.zeros((len(self.term2idx), len(doc_ls)))
        for doc_idx, doc in enumerate(doc_ls):
            for term in doc.split():
                tdm[self.term2idx[term], doc_idx] += 1
                
        return tdm
            
    
    # 특이값 분해
    def SVD(self, tdm):
        U, s, VT = randomized_svd(tdm, n_components = 6, n_iter = 20, random_state = None)
        return U, s , VT
        
    
    # 토픽별 주요 키워드 출력, 토픽이 결국 주제
    def TopicModeling(self, topic_num = 3):
        for i in range(topic_num):
            
            score = self.U[:, i:i+1].T # sorting
            sorted_index = np.argsort(-score)
            
            a = []
            for j in sorted_index[0 ,:topic_num]: # 인덱스를 가져옴
                a.append((self.idx2term[j], score[0, j].round(3)))
            
            print('{},{}'.format(i+1, a))
            
          
    
    def TermDocVectorMatrix(self, u, s, vt, topic_num):
        term_doc_mat = np.matrix(u[:, :topic_num]) * np.diag(s[:topic_num]) * np.matrix(vt[:topic_num, :])
        return term_doc_mat
        
    # 키워드를 입력했을 때 단어 벡터 반환
    def GetTermVector(self, term):
        vec1 = self.term_mat[self.term2idx[term]]
        print("{} = {}".format(term, vec1))
        return vec1
        
            
    # 문서를 입력했을 때 문서 벡터 반환
    def GetDocVector(self, doc):
        vec2 = self.doc_mat[self.doc2idx[doc]]
        print("{} = {}".format(doc,vec2))
        return vec2
        
    
   
    
    def TermVectorMatrix(self, u, s, topic_num):
        term_mat = np.matrix(u[:,:topic_num]) * np.diag(s[:topic_num])
        return term_mat
      
        
    def DocVectorMatrix(self, s, vt, topic_num):
        doc_mat = np.diag(s[:topic_num]) * np.matrix(vt[:topic_num ,:]) # 그냥 나중에 보여주기 편하라고 transpose 시켜줌
        return doc_mat.T
        
        
    def GetTermSimilarity(self, term1, term2): # 용어들 사이의 코사인 유사도 구해줘
        a = self.GetTermVector(term1)
        b = self.GetTermVector(term2)
        sim = cosine_similarity(a,b)
        print("{},{},{}".format(term1,term2,sim[0][0]))
        return sim
      
        
    def GetDocSimilarity(self, doc1, doc2):
        a = self.GetDocVector(doc1)
        b = self.GetDocVector(doc2)
        sim2 = cosine_similarity(a,b)
        print(sim2[0][0])
        return sim2
    
    def Compression(self, round_num=0):   
        print(self.term_doc_mat.round(round_num))
        print(self.tdm.round(round_num))
      
        
        
    
```

```python
doc_ls = ['바나나 사과 포도 포도 짜장면',
         '사과 포도',
         '포도 바나나',
         '짜장면 짬뽕 탕수육',
         '볶음밥 탕수육',
         '짜장면 짬뽕',
         '라면 스시',
         '스시 짜장면',
         '가츠동 스시 소바',
         '된장찌개 김치찌개 김치',
         '김치 된장 짜장면',
         '비빔밥 김치'
         ]

lsa = LSA(doc_ls, 3)
X = lsa.TDM(doc_ls)
lsa.TopicModeling(3)
print(X)
lsa.GetTermVector('포도')
lsa.GetTermVector('사과')
print('======================================')
lsa.GetTermSimilarity('포도', '사과')
lsa.GetTermSimilarity('볶음밥', '라면')
print('======================================')
lsa.GetDocSimilarity(doc_ls[0],doc_ls[2])
print('======================================')
lsa.Compression(2)
```

```python
# 시각화, 이해 안 돼도 ㅇㅋ s!!!

import numpy as np
from numpy import dot
from numpy.linalg import norm

def calc_similarity_matrix(vectors): 
    
    def cosine_similarity(a,b):
        return dot(a,b)/(norm(a)*norm(b))
    
    n_word = len(vectors)
    similarity_matrix = np.zeros((n_word, n_word)) # 단어별로 열과 행을 둬서 서로 얼마나 가까운지
    
    for i in range(n_word):
        for j in range(i, n_word):
            similarity_matrix[j,i] = cosine_similarity(vectors[i], vectors[j]).round(4)
            
    return similarity_matrix
```
